```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from sklearn.datasets import load_linnerud
from torch import nn
```

```python
# Data
features, response = load_linnerud(return_X_y=True)
X = torch.tensor(features)
y = torch.tensor(response)

criterion = nn.MSELoss()
```

```python
# Neural Network loss landscape setup
torch.manual_seed(0)
dtype = torch.float64
in_dim = 3
hidden = 3
model = nn.Sequential(
    nn.Linear(in_dim, hidden, bias=False),
    nn.ReLU(),
    nn.Linear(hidden, 1, bias=False),
).to(dtype)

# Function to compute loss given flattened weights
def loss_from_flat(w_flat):
    torch.nn.utils.vector_to_parameters(w_flat, model.parameters())
    return criterion(model(X), y)

# Get current weights flat
w0 = (
    torch.nn.utils.parameters_to_vector(model.parameters())
    .detach()
    .clone()
    .requires_grad_(False)
)

# Two random directions in weight space
d1 = torch.randn_like(w0)
d1 = d1 / torch.norm(d1)
d2 = torch.randn_like(w0)
d2 = d2 / torch.norm(d2)

# Grid over directions
grid_size = 500
alphas = np.linspace(-5.0, 5.0, grid_size)
betas = np.linspace(-5.0, 5.0, grid_size)
loss_vals = np.zeros((grid_size, grid_size))

for i, a in enumerate(alphas):
    for j, b in enumerate(betas):
        w_test = w0 + (a * d1 + b * d2)
        loss_vals[i, j] = loss_from_flat(w_test).item()
```

```python
# spin glass energy landscape setup
N = w0.numel()
p = 3
# Random interaction tensor J of shape (N, N, …, N) with p indices
# But for tractability we simplify: sample a single vector h and define H(s)= h⋅s^p
h = torch.randn(N, dtype=dtype)

# Define spherical unit vector constraint function
def H_spherical(s):
    s = s / torch.norm(s)
    return (h * (s**p)).sum().item()

# grid for p-spin
energy_vals = np.zeros((grid_size, grid_size))
for i, a in enumerate(alphas):
    for j, b in enumerate(betas):
        s = (w0 + (a * d1 + b * d2)).numpy()
        s = s / np.linalg.norm(s)
        energy_vals[i, j] = H_spherical(torch.from_numpy(s).to(dtype))
```

```python
# --- Plot side-by-side ---
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
im0 = axes[0].imshow(
    loss_vals,
    extent=[alphas[0], alphas[-1], betas[0], betas[-1]],
    origin="lower",
    #levels=300,
    cmap="turbo",
)
axes[0].set_title("NN Loss Landscape")
axes[0].set_xlabel("α (direction 1)")
axes[0].set_ylabel("β (direction 2)")
fig.colorbar(im0, ax=axes[0])

im1 = axes[1].imshow(
    energy_vals,
    extent=[alphas[0], alphas[-1], betas[0], betas[-1]],
    origin="lower",
    #levels=300,
    cmap="turbo",
)
axes[1].set_title("p-spin Spherical Energy Landscape")
axes[1].set_xlabel("α (direction 1)")
axes[1].set_ylabel("β (direction 2)")
fig.colorbar(im1, ax=axes[1])

plt.tight_layout()
plt.savefig('../plots/comparison.png')
plt.show()
```
